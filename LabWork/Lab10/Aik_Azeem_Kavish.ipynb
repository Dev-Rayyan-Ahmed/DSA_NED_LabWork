### Exercise (NaN)

```python
class Queue:
    def __init__(self, size):
        self.queue = [0]*size
        self.size = size
        self.front = -1
        self.rear = -1
    
    def isEmpty(self):
        if not self.isFull():
            return self.rear == self.front
        return False
        
    def isFull(self):
        if self.rear != -1 and self.front != -1:
            return (self.rear)% self.size == self.front
        return False
    
    def enqueue(self, item):
        if self.isFull():
            print("FULL")
            return
        if self.isEmpty():
            self.rear = 0
            self.front = 0

        self.queue[self.rear] = item
        self.rear = (self.rear + 1) % self.size
    
    def dequeue(self):

        if self.isEmpty():
            print("EMPTY")
            return -1
        
        val = self.queue[self.front]
        self.queue[self.front] = "NaN"
        self.front = (self.front + 1)% self.size
        return val

    def __str__(self):
        result = "[ "
        
        for i in range(self.size):

            result += (str(self.queue[i]))

            if i == self.front:
                result += "(f)"

            if i == self.rear:
                result += "(r)"

            result += " " 

        result += "]"

        return result

Q = Queue(4)
print("0",Q)
Q.enqueue(0)
print("1",Q)
Q.enqueue(1)
print("2",Q)
Q.enqueue(2)
print("3",Q)
Q.enqueue(3)
print("4",Q)

Q.dequeue()
Q.dequeue()
Q.dequeue()

print("Dequeued: ", Q)

Q.enqueue(0)
print("5",Q)
Q.enqueue(1)
print("6",Q)
Q.enqueue(2)
print("7",Q)
Q.enqueue(3)
# print("8",Q)
```
